```python
import pandas as pd
import requests
```

```python
resposta = requests.get(api_url)
resposta.raise_for_status()
```

```python
arquivos = resposta.json()
```

Importando todos os dados.

```python
emprestimos_total = []
```

```python
base_raw_url = 'https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/48684f508c0c9ad1d51975c72091b160dee66db4/Dia_1-Importando_dados/Datasets/dados_emprestimos/'
```

```python
for arquivo in arquivos:
    file_name = arquivo['name']  # Obtém o nome do arquivo do dicionário
    if file_name.endswith('.csv'):  # Verifica se o arquivo é um CSV
        file_url = base_raw_url + file_name  # Constrói a URL do arquivo CSV
        df = pd.read_csv(file_url)  # Lê o CSV em um DataFrame
        emprestimos_total.append(df)  # Adiciona o DataFrame à lista de DataFrames
```

```python
emprestimos_juntos = pd.concat(emprestimos_total, ignore_index=True)
```
